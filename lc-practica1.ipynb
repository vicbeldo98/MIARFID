### 1) Obtener un diccionario, que para cada categoría, muestre su frecuencia. Ordenar el resultado alfabéticamente por categoría:
C 2

DT 4

Fp 1

N 7

P 2

V 2


```python
def primer_ejercicio(cadena):
    pairs = [x for x in cadena.split(" ")]
    res = {}
    for pair in pairs:
        elements = pair.split("/")
        if not res.get(elements[1], None):
            res[elements[1]] = 0
        res[elements[1]] += 1
    
    return sorted(res.items())

res = primer_ejercicio(cadena)
# Pretty print
for i in res:
    print(i[0] + ' ' + str(i[1]) + '\n')
```

### 2) Obtener un diccionario, que para cada palabra, muestre su frecuencia, y una lista de sus categorías morfosintácticas con su respectiva frecuencia. Mostrar el resultado ordenado alfabéticamente por palabra:

. 1 Fp 1

 canta 1 V 1
 
 carne 1 N 1
 
 carnicería 1 N 1
 
 come 1 V 1
 
 de 2 P 2
 
 el 2 DT 2
 
 la 5 DT 2 N 3
 
 nevera 1 N 1
 

 perro 1 N 1
 
 y 2 C 2


```python
def segundo_ejercicio(cadena):
    pairs = [x for x in cadena.split(" ")]
    res = {}
    for pair in pairs:
        elements = pair.split("/")
        word = elements[0].lower()
        morf = elements[1]
        if not res.get(word, None):
            res[word] = [0, {}]
        res[word][0] += 1
        morf_dict = res[word][1]
        if not morf_dict.get(morf, None):
            morf_dict[morf] = 0
        morf_dict[morf] += 1
    
    return sorted(res.items())

res = segundo_ejercicio(cadena)

# Pretty print
str_res = ''
for i in res:
    str_res += i[0] + ' ' + str(i[1][0])
    freq_dict = i[1][1]
    for item in freq_dict.keys():
        str_res += ' ' + item + ' ' + str(freq_dict[item])
    str_res +='\n'

print(str_res)
```

### 3) Calcular la frecuencia de los todos los bigramas de la cadena, considerar un símbolo inicial < S > y un símbolo final < /S > para la cadena.

('DT', 'N') 4

('N', 'V') 1

('N', 'C') 2

('N', 'Fp') 1

('N', 'N') 2

('C', 'V') 1

('V', 'N') 1

('V', 'DT') 1

('P', 'DT') 2

('Fp', '< /S >') 1

('< S >', 'DT') 1
    
('C', 'P') 1

('N', 'P') 1

```python
def tercer_ejercicio(cadena):
    res = {}
    words = [x.split("/")[1] for x in cadena.split(" ")]
    words.insert(0, '<S>')
    words.append('</S>')
    for i in range(len(words)-1):
        if not res.get((words[i], words[i+1]), None):
            res[(words[i], words[i+1])] = 0
        res[(words[i], words[i+1])] +=1
    return res

res = tercer_ejercicio(cadena)

# Pretty print
str_res = ''
for i in res.keys():
    str_res += str(i) + ' ' + str(res[i]) + '\n'

print(str_res)
```
